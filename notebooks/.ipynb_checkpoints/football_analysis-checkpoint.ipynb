This notebook works through a dataset of international football matches. The goal is to explore scoring patterns, match outcomes, and historical team performance through basic data analysis and visualisation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')

# load the dataset — it sits in the data folder relative to this notebook
df = pd.read_csv('../data/results.csv')
df['date'] = pd.to_datetime(df['date'])
df['total_goals'] = df['home_score'] + df['away_score']

print('Dataset loaded successfully.')
print(df.head())
```

## Part 1 – Basic Exploration

```python
# basic dataset dimensions and date range
total_matches = len(df)
earliest_year = df['date'].dt.year.min()
latest_year   = df['date'].dt.year.max()

# count unique countries across both home and away columns
all_teams = pd.concat([df['home_team'], df['away_team']]).unique()
unique_countries = len(all_teams)

most_frequent_home = df['home_team'].value_counts().idxmax()
home_count = df['home_team'].value_counts().max()

print(f'Total matches:          {total_matches}')
print(f'Date range:             {earliest_year} to {latest_year}')
print(f'Unique countries:       {unique_countries}')
print(f'Most common home team:  {most_frequent_home} ({home_count} matches)')
```

## Part 2 – Goals Analysis

```python
avg_goals = df['total_goals'].mean()

# find the highest scoring match
highest_idx = df['total_goals'].idxmax()
highest_match = df.loc[highest_idx, ['date','home_team','away_team','home_score','away_score']]

total_home_goals = df['home_score'].sum()
total_away_goals = df['away_score'].sum()

most_common_total = df['total_goals'].value_counts().idxmax()

print(f'Average goals per match:  {avg_goals:.2f}')
print(f'\nHighest-scoring match:')
print(highest_match.to_string())
print(f'\nTotal home goals:         {total_home_goals}')
print(f'Total away goals:         {total_away_goals}')
print(f'Most common goals total:  {most_common_total}')
```

## Part 3 – Match Results and Home Advantage

```python
def label_result(row):
    if row['home_score'] > row['away_score']:
        return 'Home Win'
    elif row['home_score'] < row['away_score']:
        return 'Away Win'
    else:
        return 'Draw'

df['result'] = df.apply(label_result, axis=1)

result_counts = df['result'].value_counts()
result_pct    = (result_counts / len(df) * 100).round(1)

print('Match outcome breakdown:')
for outcome in ['Home Win', 'Away Win', 'Draw']:
    print(f'  {outcome:<12} {result_counts[outcome]:>5}  ({result_pct[outcome]}%)')

home_win_pct = result_pct['Home Win']
print(f'\nHome advantage: home teams win {home_win_pct}% of matches.')
print('This is notably higher than away wins, confirming a persistent home advantage.')
```

```python
# build an all-time wins table by crediting both home and away wins
home_wins = df[df['result'] == 'Home Win']['home_team'].value_counts()
away_wins = df[df['result'] == 'Away Win']['away_team'].value_counts()

all_wins = home_wins.add(away_wins, fill_value=0).astype(int).sort_values(ascending=False)

print('Top 10 countries by all-time wins:')
print(all_wins.head(10).to_string())
```

## Part 4 – Visualisations

```python
matplotlib.rcParams['font.family'] = 'DejaVu Sans'

# figure 1: goals per match distribution
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df['total_goals'], bins=range(0, df['total_goals'].max() + 2),
        color='steelblue', edgecolor='white', linewidth=0.6)
ax.set_xlabel('Total Goals in Match')
ax.set_ylabel('Number of Matches')
ax.set_title('Distribution of Goals per Match')
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('../data/goals_distribution.png', dpi=150)
plt.show()
print('Saved: goals_distribution.png')
```

```python
# figure 2: match outcome distribution
labels = ['Home Win', 'Away Win', 'Draw']
values = [result_counts[l] for l in labels]
colours = ['#2c7bb6', '#d7191c', '#ababab']

fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(labels, values, color=colours, width=0.55)
for bar, val in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 15,
            str(val), ha='center', va='bottom', fontsize=10)
ax.set_ylabel('Number of Matches')
ax.set_title('Match Outcome Distribution')
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('../data/outcome_distribution.png', dpi=150)
plt.show()
print('Saved: outcome_distribution.png')
```

```python
# figure 3: top 10 teams by all-time wins
top10 = all_wins.head(10).sort_values()

fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(top10.index, top10.values, color='#4c9c6b')
ax.set_xlabel('Total Wins')
ax.set_title('Top 10 Teams by All-Time Wins')
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig('../data/top10_wins.png', dpi=150)
plt.show()
print('Saved: top10_wins.png')
```

## Summary

The analysis confirms a few things that are well-known in football but interesting to verify from data:

- Home teams win more often than away teams, by a meaningful margin. The advantage holds consistently across the dataset.
- Low-scoring matches are the norm. The distribution of total goals is right-skewed, with 2 goals being the most common outcome.
- A handful of nations dominate the all-time wins table, reflecting both the longevity of their football programmes and the frequency with which they compete internationally.
